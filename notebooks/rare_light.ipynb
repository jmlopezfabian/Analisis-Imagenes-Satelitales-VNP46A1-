```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/home/gsu/ESCOM/Trabajo_Terminal/Analisis_Imagenes_VNP46A1/data_municipios/municipios_completos.csv")
data["Fecha"] = pd.to_datetime(data["Fecha"])
data_2024 = data[data["Fecha"].dt.year == 2024]
data_no_cdmx = data[~data["Municipio"].isin(["monterrey","oaxaca de juarez"])]
```

```python
alvaro = data_2024[data_2024["Municipio"] == "alvaro obregon"]
sns.lineplot(data=alvaro, x="Fecha", y="Suma_de_radianza", color="#301D82")
plt.grid(True)
plt.show()
```
